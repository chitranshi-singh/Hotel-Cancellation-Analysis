## 03. Exploring relationships through visualisations

## Create a correlation heatmap using seaborn

```python
# Create a correlation matrix using pandas

df.corr()
```

```python
df.columns
```

```python
df.loc[df['cancellation_rate'] < 20, 'lead_time_range'] = 'Low'

df.loc[(df['cancellation_rate'] >= 20) & (df['cancellation_rate'] < 40), 'cancellation_category'] = 'Mid'

df.loc[df['cancellation_rate'] >= 40, 'cancellation_category'] = 'High'
```

```python
# Create a subset

sub = df[['lead_time', 'cancellation_rate', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'kids', 
          'booking_changes', 'days_in_waiting_list', 'adr', 'required_car_parking_spaces', 'total_of_special_requests']]
```

```python
sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(9,6))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

### The lighter the orange or the darker the purple, the stronger the (linear) relationship between the two variables. Some observations from this heatmap:

#### The strongest correlation is between 'lead_time' and 'cancellation_rate'. Meaning the higher the lead time, the higher the chance of its booking being cancelled.

#### 'stays_in_weekend_nights' and 'stays_in_week_nights' has a moderately strong correlation too. Meaning the longer the guests stay on weekend nights, the longer is their stay during week nights, which is kind of to be expected.

#### There isn't really a strong negative correlation between any two variables in the heatmap above.

#### The positive correlation between 'lead_time' and 'cancellation_rate' will be worth exploring.

## Categorical Plots

### Creating a categorical variable that splits the 'cancellation_rate' column into categories

```python
# use a histogram to visualize the distribution of variable

sns.histplot(df['cancellation_rate'], bins = 20, kde = True)
```

```python
df.loc[df['cancellation_rate'] < 20, 'cancellation_category'] = 'Low'

df.loc[(df['cancellation_rate'] >= 20) & (df['cancellation_rate'] < 40), 'cancellation_category'] = 'Mid'

df.loc[df['cancellation_rate'] >= 40, 'cancellation_category'] = 'High'
```

```python
# 'lead_time' and 'cancellation_rate'

sns.set(style='ticks')
g = sns.catplot(x='lead_time', y='cancellation_rate', hue='cancellation_category', data=df)
```

#### Further to what was identified earlier, there are a few cases when despite having a high lead time, the cancellation is low. This could be because the distribution of 'cancellation_rate' is slightly bimodal in nature. However, there are many more instances of high cancellation as the lead time increases.
#### Thus, as a takeaway, the lead time could be a good predictor of cancellation.
